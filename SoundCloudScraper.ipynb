```
#@title Artist and quantity { vertical-output: true, output-height: 1, form-width: "30%", display-mode: "form" }
#@markdown Choose artist and nr of tracks

#@markdown To download press ctrl-F9

# https://github.com/Miserlou/SoundScrape.git
!wget https://i1.wp.com/www.stortz.com/wp-content/uploads/2015/03/356-d-1.jpg
!mv /content/356-d-1.jpg /content/logo.jpg

artist = "thedarkhourssessions" #@param ["iamtlabel", "codexrecordings", "g7l87f67gi", "fearloathing", "t-sessions", "andrewprojectofficial", "renato-caltagirone", "kembo", "georgi-dimov-11", "dj-spoony-1", "thedarkhourssessions"] {allow-input: true}
qty_to_rip=10 #@param {}

site='soundcloud' #@param ['mixcloud','soundcloud']
# imports and defs
!rm -r /content/sample_data
import sys, time, os, glob, cv2, itertools, fnmatch
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import clear_output, Image, HTML
from google.colab.patches import cv2_imshow
import datetime

from IPython.display import Audio as AUDIO
from IPython.display import display as DSPL
!pip install ipywebrtc
import ipywebrtc as ipyweb
from ipywebrtc import AudioStream
import ipywidgets
from ipywidgets import Image, Video, Box, Audio,Layout, Button
from bokeh.models.widgets.inputs import AutocompleteInput

# print(AUDIO.__doc__)
# print(ipywidgets.Audio.__doc__)


# list of reps to install
reps=[
        'openai/universe',
        'openai/gym',
#         'EIGENREPS',
#         'bxck75/piss-ant-pix2pix',
#         'bxck75/gallery-dl',
        'librosa/librosa',
#         'jupyter-widgets/ipywidgets',
        'bxck75/SoundScrape',
#         'bxck75/cookbook-2nd-code',
#         'bxck75/datasets',
#         'EIGENREPS',
#         'tjwei/Flappy-Turtle.',
#         'tjwei/fonttools',
#         'tjwei/blender3d_import_psk_psa',
#         'lllyasviel/sketchKeras',
#         'Mckinsey666/Anime-Face-Dataset',
#         'chenyuntc/pytorch-book',
#         'lllyasviel/style2paints',
#         'llSourcell/GANS-for-style-transfer',
#         'opencv/open_model_zoo',
#         'hindupuravinash/the-gan-zoo',
#         'corenel/GAN-Zoo',
#         'eriklindernoren/Keras-GAN',
#         'junyanz/CycleGAN',
#         'junyanz/pytorch-CycleGAN-and-pix2pix',
#         'junyanz/iGAN', #----> !wget http://efrosgans.eecs.berkeley.edu/iGAN/datasets/church_64.zip <----dataset 	outdoor_128.zip 	handbag_128.zip !!!
#         'martinarjovsky/WassersteinGAN',
#         'shaoanlu/faceswap-GAN',
#         'LantaoYu/SeqGAN',
#         'tjwei/GANotebooks',
#         'adeshpande3/Tensorflow-Programs-and-Tutorials',
#         'adeshpande3/Generative-Adversarial-Networks',
#         'diegoalejogm/gans',
#         'osh/KerasGAN',
#         'tensorflow/gan',
#         'r9y9/gantts',
#         'jayleicn/animeGAN',
#         'jayleicn/ImageNet-Training',
#         'Zardinality/WGAN-tensorflow',
#         'timsainb/Tensorflow-MultiGPU-VAE-GAN',
#         'Larox/python-moviepy-meetup',
#         'tjwei/keras-yolo3',
#         'tensorflow/moonlight'
#         'tensorflow/models',
#         'tensorflow/datasets',
#         'tensorflow/docs',
#         'tensorflow/examples',
#         'LynnHo/CycleGAN-Tensorflow-2',
#         'mnicnc404/CartoonGan-tensorflow',
#         'imatge-upc/3D-GAN-superresolution',
]
print(reps.sort())

# Gitgo class
class GitGo():
    
  def __init__(self,rep,chdir=True,path='/content/'):
    self.rep = rep
    self.chdir = chdir
    self.path = path
    !mkdir -p {self.path}
    if self.rep == 'help':
        self.help()
    self.go() 
      
  def help(self):
    print("* pulls git rep and shows files \
            * returns root path for the repository \
            * Function needs repository <user>/<repository name> combination\
            * Switch chdir and define the rootpath for the repository\
            * Use : GitGo('<rep owner>/<rep name>', <True/False>, <root path>)\
          ")
    sys.exit()
    
  def go(self):
    self.rep=self.rep.split('/')
    # change folder check
    if self.chdir ==True:
        #Switch to path
        os.chdir(self.path)
    # pull get the git rep
    os.system('git clone https://github.com/'+self.rep[0]+'/'+self.rep[1]+'.git')
    # Set the return value for rep rootpath
    self.PATH=self.path+self.rep[1]
    # show imported files
    os.system('ls ' + self.PATH)

  def __repr__(self):
    return self.PATH # +'/'


def get_other_reps(reps):
    GitUsers=[]
    for u in reps:    
        username=u.split('/')[0]
        GitUsers.append(username)
        GHUSER=username
        !curl 'https://api.github.com/users/{GHUSER}/repos?per_page=100' | grep -o 'git@[^"]*' 

    return list(set(GitUsers))

def recursive_glob(treeroot, pattern):
    results = []
    for base, dirs, files in os.walk(treeroot):
        goodfiles = fnmatch.filter(files, pattern)
        results.extend(os.path.join(base, f) for f in goodfiles)
    return results

other_reps_list = []
rep_paths=[]
# install all reps in the list
for rep in reps:
    name=rep.split('.')[0]
    print(name)
    path='/content/installed_resp'
    G=GitGo(name,path=path)
    R=get_other_reps([name])
    rep_paths.append([name,G])
    other_reps_list.append([R[0],name])
    
# mylist = get_other_reps(reps)
# print(other_reps_list)
# print(rep_paths)
# recursive_glob('/content/','requirements*.txt')


# print(rep_paths)
paths= recursive_glob('/content/','requirements*.txt')
# installing rep requirements
!pip install -r {paths[0]}
%cd /content/installed_resp/SoundScrape
!python setup.py install
# run test
# !bash /content/installed_resp/SoundScrape/test.sh
# clear the long output
clear_output()
  
```
